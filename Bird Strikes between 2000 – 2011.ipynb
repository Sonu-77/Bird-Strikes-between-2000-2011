```python
# used dictionary to fill the Na value with 'Nothing' 
data_frame.fillna({'Effect: Impact to flight': 'Nothing'}, inplace=True) 

#used dictionary to fill the Na value with 'None'
data_frame.fillna({'Conditions: Precipitation': 'None'},inplace=True)   

# used dictionary to fill the Na value with 'No Remarks' 
data_frame.fillna({'Remarks': 'No Remarks'},inplace=True) 

# used dictionary to fill the Na value with 'Unknown'
data_frame.fillna({'Airport: Name':'Unknown'},inplace=True) 

# used dictionary to fill the Na value with 'Unknown'
data_frame.fillna({'Origin State':'Unknown'},inplace=True) 

# used dictionary to fill the Na value with 'Unknown'
data_frame.fillna({'Aircraft: Airline/Operator':'Unknown'},inplace=True) 
```

```python
# used the 'for loop' to make an list to jump over the columns
for column in ['Aircraft: Type', 'Altitude bin',
                'When: Phase of flight','Aircraft: Number of engines?',
               'Pilot warned of birds or wildlife?', 'Is Aircraft Large?', 'Feet above ground']:
    mode_value = data_frame[column].mode()[0]
data_frame[column] = data_frame[column].fillna(mode_value)
```

```python
# using the if-else function to replace the missing cells in the 'Wildlife: Number struck'
def fill_missing_struck(row):
    # If the 'Wildlife: Number struck' column is not NaN, just return its value
    if pd.notna(row['Wildlife: Number struck']):
        return row['Wildlife: Number struck']
    # If 'Wildlife: Number struck' is NaN,  return based on 'Wildlife: Number Struck Actual'
    else:
        if row['Wildlife: Number Struck Actual'] > 100:
            return 'Over 100'
        elif 10 < row['Wildlife: Number Struck Actual'] <= 100:
            return '11 to 100'
        else:
            return '2 to 10'
data_frame['Wildlife: Number struck'] = data_frame.apply(fill_missing_struck, axis=1)
```

```python
#droping the rows with most of the  missing data.
data_frame.dropna(subset=['Aircraft: Number of engines?','FlightDate'], inplace=True) 
```

# Top 10 Birds Struck on the following Airport 

```python
Airport_Name_gb_agg=data_frame.groupby('Airport: Name')['Wildlife: Number Struck Actual'].sum().to_frame().reset_index()
```

```python
Sorted_Airport_Name=Airport_Name_gb_agg.sort_values(by='Wildlife: Number Struck Actual',ascending=False)
```

```python
Top_10_Airpor_Name=Sorted_Airport_Name.head(10)
Top_10_Airpor_Name
```

```python
plt.figure(figsize=(20, 10))

sns.set(style="white") # background color

# seaborn bar plot
bar=sns.barplot(data=Top_10_Airpor_Name, x='Airport: Name', 
                y='Wildlife: Number Struck Actual', 
                edgecolor='black', color='green',lw=2)

# Set labels 
plt.xlabel('Birds Struck', fontweight='bold')
plt.ylabel('Airport Names', fontweight='bold')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)  

#title
plt.title('Top 10 Birds Struck On Airports',
          fontweight='bold',fontsize=25,color='#FBA004')

#axislabels
bar.tick_params(axis='x',colors='#FB055F') # coloring x-axis tick labels
bar.tick_params(axis='y',colors='#17202A') # coloring y-axis tick labels

# Adding value labels on top of each bar

for i, height in enumerate(Top_10_Airpor_Name['Wildlife: Number Struck Actual']):
    plt.text(i,height -(0.07*height),str(height),ha='center',
             va='top',fontweight='bold',color='#FFBF00')

plt.show()
```

#### INSIGHT : 

-  The highest no. of  Birds strucks on the Airports was  DALLAS/FORT WORTH INTL ARPT with an count of "2932", as the barplot show the other Airports with there birds struck's count.

# Pie Chart to show the Birds struck by Aircraft Models

```python
Aircraft_model=data_frame.groupby('Aircraft: Make/Model')['Wildlife: Number Struck Actual'].sum().to_frame().reset_index()
```

```python
Top_10_model_struck=Aircraft_model.sort_values(by='Wildlife: Number Struck Actual',ascending=False).head(10)
```

```python
Top_10_model_struck
```

```python
plt.figure(figsize=(25,10))

# background color
sns.set(rc={'figure.facecolor':'#FFFFFF'}) # to change the default background of the canvas.

explodes=[0.1]*10 # individualy giving the exploding size for the sectors.

colors=['#A51C2C','#FBA004','#52BE80',
        '#D32F2F','#6A1B9A','#6D4C41','#E91E63',
        '#BDBDBD','#DCE775','#29B6F6']

plt.pie(Top_10_model_struck['Wildlife: Number Struck Actual'],
        labels=Top_10_model_struck['Aircraft: Make/Model'],
        radius=0.9,explode=explodes,colors=colors,autopct='%0.2f%%',pctdistance=0.7,
        shadow=True,labeldistance=1.1,startangle=90,
        textprops={'fontsize':14,'fontweight':'bold','color': 'k'},
        wedgeprops={'linewidth':3,'edgecolor':'#000000'})

plt.title('Distribution of Wildlife Struck By Individual Aircraft Models', 
          fontweight='bold',fontsize=15,color='#0000CC')

plt.show()
```

#### INSIGHT :

-
The pie chart shows the models with the highest no. of birds struck by each Aircraft models, as the highest no. of birds struck by the model is CL-RJ100/200 with count of "5878". and share 17.87% of the top 10 models in the chart.
